```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("high_dimensional_dataset.csv")  

# Preview the dataset
df.head()
```

```python
# Step 3: Apply PCA
X = df.drop('class', axis=1)
y = df['class']
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
df_pca['class'] = y
```

```python
# Step 4: Visualize
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_pca, x='PC1', y='PC2', hue='class', palette='Set2', s=70)
plt.title('PCA - 2D Projection')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df_pca.to_csv("reduced_dataset_pca.csv", index=False)
print("✅ Reduced dataset saved as 'reduced_dataset_pca.csv'")
```
